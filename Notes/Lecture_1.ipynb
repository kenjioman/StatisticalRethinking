## Key Ideas/ Summary
### Models vs Statistical Tests:
* Models are potentially powerful tools to help us better understand the world better, but with their power, comes the potential to abuse it, so beware.
 - Statistical testing is just using a set of models previously designed (with their sets of assumptions) -- instead of using/ memorizing pre-canned models, let's understand the process of building and evaluating models directly.
 - Hypotheses lead to process models lead to statistical models -- remember, the same statistical model can be explained by contradictory hypotheses/ process models.
 
### Framework for model building/ vetting:
* Bayesian data analysis
 - at the core, just count all the ways data can happen, according to the assumptions
 - Models with more ways that are consistent with the data are more plausible
* Multilevel models
 - Allows one one build more complex models (models within models), to better capture the complexity of modern scientific problems.
* Model comparisons
 - Compare different models to see which is better
 - Recognize when something has gone wrong

### Small vs large world:
* Worry about both what set of assuptions went into the construction of the small world of the model, and how that relates/ reflects the real "large" world.

### Model building strategy:
1. Design the model (data story)
2. Condition on the data (update)
3. Evaluate the model (critique)

1 leads to 2 leads to 3, leads to 1.

### Bayesian philosophy -- against tests
* "tests" are specialized pre-made "procedures"
* Most developed in early 20th century, fragile, and eclipsed with more recent tools
 * Usually developed in randomized trials with large effects, where measurement problems are solved
* Don't believe: **Purpose of a statistical procedure is to falsify a hypothesis**.
 - Fine to say purpose of research paradigm is to falsify hypothesis, but that is *not* the purpose of a statistical test


#### Example from molecular evolution:
* Statistical models are *not* hypothesis
* 3 levels in scientific process

 * Hypothesis (tends to be/ start out more vague)
  - Evolution is neutral
  - Selection matters

* Process Models (a mathematical instantiation of the hypothesis)
  - Neutral equillibrium
  - Constant Selection
  - Fluctuating Selection

* Statistical Models (to test to process model with real data)

* **Same hypothesis can create several process models. Several process models can create statistical models. Same statistical model can arise from different process models, even though the underlying hypotheses are different/ competing! (See 21:12 in the video for chart/ description)**.

### Failure of falsification (philosophy of Karl Popper)
* Null models not unique
* Should falsify *explanatory* model, not the *null* model

### What is Baysian Data Analysis?
* The **original** way of doing data analysis
* Use probability to describe uncertanty
 * Extends ordinary logic (true/false) to continuous plausibility
* Computationally difficult, so needed Markov Chain Monte Carlo (MCMC)
* Used to be controversial, but basically all stats departments are now Bayesian
 * Ronald Fisher (early 20th century): Bayesian analysis "must be wholly rejected".
 
Core idea:
* Count all the ways data can happen, according to the assumptions
* Assumptions with more ways that are consistent with data are more plausible.

### Multilevel Models
* Models with *multiple levels* of uncertainty
 * replace parameters with models (have models in models)
* Lots of uses (common ones):
 * Repeat/ imbalanced sampling
 * Study variation
 * Avoid averaging
 * Phylogenetics, factor and path analysis, networks, spatial models
* Natural Bayesian strategy

### Model comparison
* Compare multiple models
* Instead of falsifying a null model, compare **meaningful** models.
* Basic problem
 - Overfitting -- "most important phenomenom in statistics" is model learns the data too well, leads to poor statistics (scientists frequently do this too much, while industry frequently handles this better).
 - Causal inference
* Must distinguish prediction from inference.

### Small and Large Worlds
* Sensu L.J. Savage (1954)
 - Small world: The world of the model's assumptions.  Bayesian golems are optimal, in the small world
 - Large world: The real world.  No guarantee of optimality for any kind of model.
* We have to worry about both worlds.

### Garden of forking data
* The future:
 - Full of branching paths
 - Each choice closes some
* The data:
 - Many possible events
 - Each observation eliminates some
 
Example of bag with blue/white marbels.  Show all possible paths/ possibilities with the assumed number of blue/white marbels.

* Bayesian updating -- with new information, update our expectations.
* Bayesian inference is good in that it can more easily use all data (information of a different type)

### Counts to plausibility
* Basis of applied probability
 - Things that can happen more ways are more plausible
* Plausibility is *probability*: Set of non-negative real numbers that sum to one
 - Probability theory is just a set of shortcuts for counting
